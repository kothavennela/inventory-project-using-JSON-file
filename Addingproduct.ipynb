**LOADING JSON**

```python
import json
fo = open("deletingnewrecord.json",'r')
read = fo.read()
fo.close()

productrecord = json.loads(read)
```

**ADDING NEW ITEM TO RECORD**

```python
prodid = str(input("Enter product id:"))
prodname = str(input("Enter name:"))
prodcategory = str(input("Enter category:"))
prodprice = int(input("Enter price:"))
prodquantity =  int(input("Enter quantity:"))
productdiscount =  int(input("Enter  discount:"))
productrecord[prodid] = {'itemname': prodname, 'prodcategory':prodcategory , 'prodprice':prodprice,'prodquantity':prodquantity,'productdiscount':productdiscount}
print("new record entered successfully")
import json
js = json.dumps(productrecord)

fo = open("productrecord.json",'w')
fo.write(js)
fo.close()

```
